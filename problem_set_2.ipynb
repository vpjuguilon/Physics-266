```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import j1  # Bessel function of the first kind

# Constants
wavelength = 0.550e-3  # Wavelength in mm (0.550 micron)
a = 10  # Aperture radius in mm (1 cm)
f = 30  # Focal length in mm (3 cm)
n = 1.53  # Refractive index

# Define grid for the x-y plane
grid_size = 1000
extent_aperture = 12  # Aperture extent in mm
extent_pattern = 0.001  # Diffraction pattern extent in mm
x_ap = np.linspace(-extent_aperture, extent_aperture, grid_size)
y_ap = np.linspace(-extent_aperture, extent_aperture, grid_size)
X_ap, Y_ap = np.meshgrid(x_ap, y_ap)
R_ap = np.sqrt(X_ap**2 + Y_ap**2)

# Define grid for diffraction pattern
x = np.linspace(-extent_pattern, extent_pattern, grid_size)
y = np.linspace(-extent_pattern, extent_pattern, grid_size)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)

# Aperture function
aperture = np.where(R_ap <= a, 1, 0)

# Wave number
k = 2 * np.pi * n / wavelength

# Fraunhofer diffraction pattern (normalized intensity)
epsilon = 1e-12  # Avoid division by zero at the center
arg = (k * a * R) / f
intensity = np.where(arg > epsilon, (2 * j1(arg) / arg)**2, 1)

# Normalize intensity to I0 = 1
intensity /= np.max(intensity)

# Plot aperture
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(aperture, extent=[-extent_aperture, extent_aperture, -extent_aperture, extent_aperture], cmap='gray')
plt.colorbar(label="Amplitude")
plt.title('Circular Aperture')
plt.xlabel('x (mm)')
plt.ylabel('y (mm)')
plt.xlim([-extent_aperture, extent_aperture])
plt.ylim([-extent_aperture, extent_aperture])

# Plot diffraction pattern
plt.subplot(1, 2, 2)
contour = plt.contourf(X, Y, intensity, levels=np.arange(0, 1.05, 0.05), cmap='inferno')
plt.colorbar(label="Normalized Intensity")
plt.title('Fraunhofer Diffraction Pattern')
plt.xlabel('x (mm)')
plt.ylabel('y (mm)')
plt.xlim([-extent_pattern, extent_pattern])
plt.ylim([-extent_pattern, extent_pattern])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
wavelength = 0.550e-3  # Wavelength in mm (0.550 micron)
a = 20  # Aperture height in mm (2 cm)
b = 10  # Aperture width in mm (1 cm)
f = 30  # Focal length in mm (3 cm)
n = 1.53  # Refractive index

# Define grid for the x-y plane
grid_size = 1000
extent = 0.001  # Extent in mm for display
x = np.linspace(-extent, extent, grid_size)
y = np.linspace(-extent, extent, grid_size)
X, Y = np.meshgrid(x, y)

# Wave number
k = 2 * np.pi * n / wavelength

# Fraunhofer diffraction pattern (normalized intensity)
epsilon = 1e-12  # Small value to avoid division by zero
arg_x = (k * b * X) / (2 * f)
arg_y = (k * a * Y) / (2 * f)

# Compute sinc pattern
intensity = (
    (np.sinc(arg_x / np.pi) ** 2) * (np.sinc(arg_y / np.pi) ** 2)
)

# Normalize intensity to I0 = 1
intensity /= np.max(intensity)

# Plot aperture and diffraction pattern
plt.figure(figsize=(12, 5))

# Aperture plot
plt.subplot(1, 2, 1)
aperture = np.where((np.abs(X) <= b / 2) & (np.abs(Y) <= a / 2), 1, 0)
plt.imshow(aperture, extent=[-extent, extent, -extent, extent], cmap='gray')
plt.colorbar(label="Amplitude")
plt.title('Rectangular Aperture')
plt.xlabel('x (mm)')
plt.ylabel('y (mm)')
plt.xlim([-extent, extent])
plt.ylim([-extent, extent])

# Diffraction pattern plot
plt.subplot(1, 2, 2)
contour = plt.contourf(X, Y, intensity, levels=np.arange(0, 1.05, 0.05), cmap='inferno')
plt.colorbar(label="Normalized Intensity")
plt.title('Fraunhofer Diffraction Pattern')
plt.xlabel('x (mm)')
plt.ylabel('y (mm)')
plt.xlim([-extent, extent])
plt.ylim([-extent, extent])

plt.tight_layout()
plt.show()
```
